En este notebook se muestra el proceso de ajuste de una recta para los datos de:
- Diferencia de potencial,
- Frecuencia,

usando el comando numpy.polyfit del paquete numpy.

```python
# =============================================
# Se importan paquetes y funciones de utilidad.
# =============================================

import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sp

e = 1.602e-19

def b(x,m,c):
  """
  Ecuaci칩n de recta.
  """
  return m*x + c

def recta_de_mejor_ajuste(x,y):
  """
  Funci칩n que grafica una recta de mejor ajuste para los datos de entrada, en
  este caso, frecuencia en el eje X y diferencia de potencial en el eje Y.
  Las listas deben estar en orden entre ellas, por ejemplo, al valor de x[0] le
  corresponde y[0].

  Par치metros:
    x: Datos de entrada en el eje x (frecuencia).
    y: Datos de entrada en el eje y (diferencia de potencial).
  """
  # ============
  # Graficaci칩n.
  # ============

  # Se define una recta de ajuste para los puntos de entrada.
  popt,_ = sp.curve_fit(b,x,y)

  plt.scatter(x,y)
  plt.plot(x,b(x,popt[0],popt[1]),'r-',label=str(popt[0])+'x'+str(popt[1]))
  plt.xlabel('Frecuencia')
  plt.ylabel('Diferencia de potencial')
  plt.legend()
  plt.show()

  print(f'El valor de la pendiente es: {popt[0]} y su unidad de medida es [V/Hz]')
  print(f'El valor para la constante de Planck es: {popt[0]*e}[J/Hz]')
  return
```

```python
# =================================================================
# Se definen listas con los objetos en orden para ajustar la recta.
# =================================================================

dif_pot = np.array([1.436,1.678,1.737,1.844,2.383])

frec_no_escal = np.array([4.292,5.004,5.213,5.623,6.645])
frec_escal = frec_no_escal*(10**14)

# ==============================================
# Se calcula la recta y se muestra la pendiente.
# ==============================================

recta_de_mejor_ajuste(frec_escal,dif_pot)
```
