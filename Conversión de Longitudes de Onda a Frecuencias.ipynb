En este notebook se mostrará la conversión de ciertas longitudes de onda a sus valores de frecuencia asociados usando la relación:
\begin{equation}
  \lambda\nu=c\;.
\end{equation}
Los valores para las longitudes de onda son el promedio de los límites superior e inferior de los intervalos de definición dados por este link:

https://es.wikipedia.org/wiki/Espectro_visible

Se considera un valor de $c=3.000\cdot10^{17}[nm\cdot s^{-1}]$ para el cálculo.

```python
# ==============================================================================
# Se definen listas que contienen los valores de longitud de onda de los colores
# de interés y con las etiquetas de colores para su identificación, además de la
# constante de la velocidad de la luz.
# ==============================================================================

longitudes_de_onda = [699.0,599.5,575.5,533.5,451.5] # Valores en [nm]
etiquetas_de_color = ['rojo','naranjo','amarillo','verde','azul']
c = 3.000*(10**17)
```

```python
# =====================================================================
# Se realiza el cálculo y se guardan los resultados en una lista vacía.
# =====================================================================

frecuencias = []

for i in range(0,5):
  frecuencias.append(c/longitudes_de_onda[i])
```

```python
# ==========
# Resultado.
# ==========

for i in range(0,5):
  print(f'Color:            {etiquetas_de_color[i]}.')
  print(f'Longitud de onda: {longitudes_de_onda[i]}[nm].')
  print(f'Frecuencia:       {frecuencias[i]}[Hz]')
```

Así, se obtienen los valores para frecuencia:
- Rojo:     $4.292\cdot10^{14}[Hz]$
- Naranjo:  $5.004\cdot10^{14}[Hz]$
- Amarillo: $5.213\cdot10^{14}[Hz]$
- Verde:    $5.623\cdot10^{14}[Hz]$
- Azul:     $6.645\cdot10^{14}[Hz]$
